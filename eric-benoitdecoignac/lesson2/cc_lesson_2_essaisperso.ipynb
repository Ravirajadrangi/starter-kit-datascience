```python
# Given an array of ints, return True if one of the first 4 elements
# in the array is a 9. The array length may be less than 4.
def array_front9_sol1(nums):
    for num in nums[:4]:
        if num == 9:
            return True
        else:
            return False
        
def array_front9_sol2(nums):
    reponse = False
    for num in nums[:4]:
        if num == 9:
            reponse = True
    return reponse
```

```python
def last2_sol1(string):
    end_substring = string[-2:]
    count = 0
    for i in range(len(string) - 2):
        if string[i : i+2] == end_substring:
            count += 1
    return count
```

```python
# Given a string, return the count of the number of times
# that a substring length 2 appears  in the string and also as
# the last 2 chars of the string, so "hixxxhi" yields 1 (we won't count the end substring).
def last2(string):
    return sum([1 if string[i:i+2]==string[-2:] else 0 for i in range(len(string)-2)])
```

```python
string = "hihihixxxxhi"
```

```python
print(last2("hihihixxxxhi"))
```

```python
def last2_while(string):
    end_substring = string[-2:]
    count = 0
    i = 0
    while i < len(string) - 2:
        if string[i : i+2] == end_substring:
            count += 1
        i += 1
    return count
```

```python
def pigLatin_oneshot(text):
    return ' '.join([word[1:] + word[0] +'ay' for word in text.split() if word.isalnum()]).capitalize()
```
