```python
host = 'http://base-donnees-publique.medicaments.gouv.fr'
end='/index.php#result'
url = host + end
```

```python
def soupify():
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser') 
    return soup 
```

```python
soupify().find_all(class_='gras' == 'btnSubst')
```

```python
def poster(recherche):
    params = {'page':1, 'affliste':0, 'affNumero':0, 'isAlphabet':0, 'inClauseSubst':0, 'nomSubstances': '', 'typeRecherche':0, 'choixRecherche':'medicament', 'paginationUsed':0, 'txtCaracteres':recherche, 'btnMedic.x':15, 'btnMedic.y':10, 'btnMedic':'Rechercher', 'radLibelle':1, 'txtCaracteresSub': '', 'radLibelleSub':4}
    postresult = requests.post(url, data=params)
    soup = BeautifulSoup(postresult.text)
    return soup
```

```python
poster('ibuprofene').find_all('td', {"gras"})
```
