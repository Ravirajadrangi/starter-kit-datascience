```python
def strip_corse(val):
    if type(val) == int:
        return val
    if val == 'nan':
        return -1
    return (re.sub('(2A|2B)', '20',val))

medecin_departement['SPECIALITE'] = medecin_departement['SPECIALITE'].apply(strip_corse)
```
