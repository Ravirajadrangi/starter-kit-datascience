```python
#### MON CODE ####

import requests
import bs4
import os


### Pour l'url de chaque page : 

def Chaque_Page(nombre, marque) : 
    if marque =='ACER' :
        url = "http://www.cdiscount.com/search/10/ordinateurs.html?TechnicalForm.SiteMapNodeId=0&TechnicalForm.DepartmentId=10&TechnicalForm.ProductId=&hdnPageType=Search&TechnicalForm.SellerId=&TechnicalForm.PageType=SEARCH_AJAX&TechnicalForm.LazyLoading.ProductSheets=False&NavigationForm.CurrentSelectedNavigationPath=0&FacetForm.SelectedFacets.Index=0&FacetForm.SelectedFacets.Index=1&FacetForm.SelectedFacets.Index=2&FacetForm.SelectedFacets%5B2%5D=f%2F6%2Facer&FacetForm.SelectedFacets.Index=3&FacetForm.SelectedFacets.Index=4&FacetForm.SelectedFacets.Index=5&FacetForm.SelectedFacets.Index=6&FacetForm.SelectedFacets.Index=7&GeolocForm.ConfirmedGeolocAddress=&SortForm.SelectedNavigationPath=&ProductListTechnicalForm.Keyword=ordinateurs&page="+str(nombre)+"&_his_"
    elif marque == 'DELL' :
        url = "http://www.cdiscount.com/search/10/ordinateurs.html?TechnicalForm.SiteMapNodeId=0&TechnicalForm.DepartmentId=10&TechnicalForm.ProductId=&hdnPageType=Search&TechnicalForm.SellerId=&TechnicalForm.PageType=SEARCH_AJAX&TechnicalForm.LazyLoading.ProductSheets=False&NavigationForm.CurrentSelectedNavigationPath=0&FacetForm.SelectedFacets.Index=0&FacetForm.SelectedFacets.Index=1&FacetForm.SelectedFacets.Index=2&FacetForm.SelectedFacets%5B2%5D=f%2F6%2Fdell&FacetForm.SelectedFacets.Index=3&FacetForm.SelectedFacets.Index=4&FacetForm.SelectedFacets.Index=5&FacetForm.SelectedFacets.Index=6&FacetForm.SelectedFacets.Index=7&GeolocForm.ConfirmedGeolocAddress=&SortForm.SelectedNavigationPath=&ProductListTechnicalForm.Keyword=ordinateurs&page="+str(nombre)+"&_his_"
    return url


### On va chercher les url de chaque ordinateur :

def Recup_prix_totaux_par_page (url) : 
    Prix_avant_promo_Par_Page = 0.0
    Prix_apres_promo_Par_Page = 0.0
    
    R = requests.get(url)
    soup = bs4.BeautifulSoup(R.content, 'html.parser')
    
    a = soup.find_all("div", class_="prdtBloc")

    for i in range(len(a)) : # on boucle sur a pour recuperer tous les prix de toutes les vignettes
        PRIX_APRES_PROMO_0 = int(a[i].find(class_="prdtPrice").text.split(u"\u20AC")[0])
        PRIX_APRES_PROMO_1 = int(a[i].find(class_="prdtPrice").text.split(u"\u20AC")[1])
        PRIX_APRES_PROMO = float(PRIX_APRES_PROMO_0) + float(PRIX_APRES_PROMO_1)/100
        
        try :
            PRIX_AVANT_PROMO_0 = int(a[i].find(class_="prdtPrSt").text.split(",")[0].replace(u'','').replace(' ',''))
            PRIX_AVANT_PROMO_1 = int(a[i].find(class_="prdtPrSt").text.split(",")[1].replace(u'','').replace(' ',''))
            PRIX_AVANT_PROMO = float(PRIX_AVANT_PROMO_0) + float(PRIX_AVANT_PROMO_1)/100
            
        except AttributeError:
            PRIX_AVANT_PROMO = PRIX_APRES_PROMO
        except ValueError:
            PRIX_AVANT_PROMO = PRIX_APRES_PROMO
        
        #print 'ordinateur ',i,' : ', PRIX_AVANT_PROMO, ' VERSUS ', PRIX_APRES_PROMO
        Prix_avant_promo_Par_Page = Prix_avant_promo_Par_Page + PRIX_AVANT_PROMO
        Prix_apres_promo_Par_Page = Prix_apres_promo_Par_Page + PRIX_APRES_PROMO
    
    #return [ round ( (Prix_avant_promo_Par_Page-Prix_apres_promo_Par_Page)/Prix_avant_promo_Par_Page, 5) , len(a)]
    return [Prix_avant_promo_Par_Page, Prix_apres_promo_Par_Page, len(a)]
    
def Etude (nombre_de_page, marque) : 
    #prix_avant_total = 0.0
    #prix_apres_total = 0.0
    P = 0.0
    quantite_totale = 0.0
    url=''
    for i in range(1, nombre_de_page + 1) : 
        url = Chaque_Page(i,marque)
        #prix_avant_total = prix_avant_total + Recup_prix_totaux_par_page(Chaque_Page(i,marque))[0]
        #prix_apres_total = prix_apres_total + Recup_prix_totaux_par_page(Chaque_Page(i,marque))[1]
        quantite_totale = quantite_totale + Recup_prix_totaux_par_page(Chaque_Page(i,marque))[1]
        print '---------------'
        print 'page ', i,' de ', marque ,': OK'
        print 'somme avant Promo / page : ', Recup_prix_totaux_par_page(url)[0]
        print 'somme apres Promo / page : ', Recup_prix_totaux_par_page(url)[1]
        #print 'quantite / page : ', Recup_prix_totaux_par_page(url)[2]
        pourcentage = round( (Recup_prix_totaux_par_page(url)[0]-Recup_prix_totaux_par_page(url)[1]) / Recup_prix_totaux_par_page(url)[0] *100, 2)
        print '% de remise / page : ', pourcentage, '%'
        P = P + pourcentage
    return P/nombre_de_page

Acer = Etude(8, 'ACER') 
Dell = Etude(8, 'DELL')

print Acer
print Dell 

if Acer > Dell : 
    print "============================================="
    print "===== Acer fait plus de remise que Dell ====="
    print "============================================="
else : 
    print "============================================="
    print "===== Dell fait plus de remise que Acer ====="
    print "============================================="
```
