```python
import re
import requests
import bs4
import os
from pandas import *

```

```python
### Crawling

#Ibuprofene levothyroxine
# nom, marque, dosage, forme galleique

# url ddu prof : http://base-donnees-publique.medicaments.gouv.fr/index.php#result 
# Mais l'utiliser sur chrome avec la requete POST et pas la GET

url_part = 'https://www.vidal.fr/recherche/index/q:levothyroxine/'
result = requests.get(url_part)
soup = bs4.BeautifulSoup(result.content, 'html.parser')


A= soup.find_all('a', class_='item_title')

#for i in range(len(A)):
#    print '\ni=',i, '\n', A[i]
# nous dit que les noms réguliers vont jusqu'à 27. Methode non automatique, mais bon...

nom = []
dosage = []
laboratoire = []
forme = []

for i in range(1,27):

    # recup du nom
    texte = str(A[i].text.split('>')) 
    NOM = texte.split(" ")[0]
    NOM = NOM[3:]
    nom.append(NOM)

    texte.replace("\\xe9","e") 
    print texte
    
    if re.search('(\d{2,4})', texte.split(" ")[1]) :
        dosage.append(str(texte.split(" ")[1]+" "+texte.split(" ")[2].replace('\\xb5', 'micro')))
        laboratoire.append('Pas d\'info')
        
        liste = texte.split(" ")[3:]
        forme_mot = ''
        for mot in liste :
            mot.replace("\\xe9","e") # ne marche pas
            forme_mot = forme_mot +' '+ mot
        forme_mot.replace("\\xe9","e")
        forme.append(forme_mot[:-2])
        
    elif re.search('(\d{2,4})', texte.split(" ")[2]) :
        dosage.append(str(texte.split(" ")[2]+" "+texte.split(" ")[3].replace('\\xb5', 'micro')))
        laboratoire.append(str(texte.split(" ")[1]))
        liste = texte.split(" ")[4:]
        forme_mot = ''
        for mot in liste :
            forme_mot = forme_mot +' '+ mot
        forme_mot.replace("\\xe9","e")
        forme.append(forme_mot[:-2])
        
    else :
        dosage.append('Pas d\'info')
        if re.search('([A-Z]+])', texte.split(" ")[1]) :
            laboratoire.append(str(texte.split(" ")[1]))
            liste = texte.split(" ")[4:]
            forme_mot = ''
            for mot in liste :
                forme_mot = forme_mot +' '+ mot
            forme_mot.replace("\\xe9","e")
            forme.append(forme_mot[:-2])
        else : 
            laboratoire.append('Pas d\'info')
            liste = texte.split(" ")[2:]
            forme_mot = ''
            for mot in liste :
                forme_mot = forme_mot +' '+ mot
            forme_mot.replace("\\xe9","e")
            forme.append(forme_mot[:-2])    
    
tab = pandas.concat([DataFrame(nom), DataFrame(laboratoire), DataFrame(dosage), DataFrame(forme)], axis=1)
tab.columns = ['nom', 'laboratoire', 'dosage', 'forme']
print tab
    
#regex pour le nom des médicaments de mon site :   LEVOTHYROXINE\s([\w\s]+)\s(\d{1,4})\s([\w\s]+), ([\w\s]+)    
    
# BUG sur EUTHYRAL : le 'cp' n'est pas mis dans le tableau
```
